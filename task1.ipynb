Q1 Perform Data cleaning on given dataset, other than missing value analysis 

removing irrelevant data

```python

not_used = ['Unnamed: 0', 'Flag', 'Club Logo', 'Real Face', 'Jersey Number','Special','Photo','Special','Work Rate','Body Type']
df = df.drop(columns=not_used)

```

```python
df.shape
```

Removing duplicates in ID column if any

```python
df.drop_duplicates(subset='ID', keep='first', inplace=True)
```

```python
df.shape
```

Formatting Joined column to date time format for better timeseries data analysis if required

```python
df['Joined'] = pd.to_datetime(df['Joined'], format='%b %d, %Y')
df['Joined'] = pd.to_datetime(df['Joined']).dt.strftime('%d-%m-%Y')


```

```python
# Extract the year from the 'Contract Valid Until' column
df['Contract Valid Until'] = df['Contract Valid Until'].str.extract(r'(\d{4})', expand=False)

# Convert the result to numeric
df['Contract Valid Until'] = pd.to_numeric(df['Contract Valid Until'], errors='coerce')

```

Q2 Demonstrate possible missing value analysis approaches using given dataset.

```python
df.isnull().sum()
```

replacing null values in CLUB column to free agent as player not part of a club

```python
df['Club'].fillna("Free Agent", inplace=True)
df['Loaned From'].fillna("Not Loaned", inplace=True)
```

```python
df.shape
```

```python
df.dropna(subset=['Height'], inplace=True)
```

```python
df.shape
```

```python
df['Position'].value_counts()
```

```python
# Define the list of columns to update
columns_to_update = ['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW', 'LAM', 'CAM', 'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM',
                     'LWB', 'LDM', 'CDM', 'RDM', 'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB']

# Conditionally set values to 0 when 'Position' is 'GK'
df.loc[df['Position'] == 'GK', columns_to_update] = 0
```

```python

df.isnull().sum()
```

```python
df.dropna(subset=['LB'], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.head(50)
```

```python
df['Wage'] = df['Wage'].str.replace('€', '').str.replace('M', 'e6').str.replace('K', 'e3').astype(float)
df['Value'] = df['Value'].str.replace('€', '').str.replace('M', 'e6').str.replace('K', 'e3').astype(float)
df = df[~((df['Wage'] == 0) & (df['Value'] == 0) & (df['Club'] == 'Free Agent'))]
```

```python
df['Release Clause'] = df['Release Clause'].str.replace('€', '').str.replace('M', 'e6').str.replace('K', 'e3').astype(float)
mean_release_clause = df['Release Clause'].mean()
df['Release Clause'].fillna(mean_release_clause, inplace=True)
```

```python
# Convert 'Contract Valid Until' to numeric
df['Contract Valid Until'] = pd.to_numeric(df['Contract Valid Until'], errors='coerce')

# Replace missing 'Joined' values with 'Contract Valid Until' - 1
df['Joined'] = df.apply(lambda row: row['Contract Valid Until'] - 1 if pd.isna(row['Joined']) else row['Joined'], axis=1)

```

Q3. Try to create a new features (columns) from existing features (columns) by aggregating the features as part of feature engineering process.

creating new column Potential_Difference by subtracting Overall from Potential to know how much more rating a player can still get

```python
df['Potential_Difference'] = df['Potential'] - df['Overall']
```

```python
def convert_height_to_inches(height):
    feet, inches = map(int, height.split("'"))
    total_inches = (feet * 12) + inches
    return total_inches

df['Height(Inch)'] = df['Height'].apply(convert_height_to_inches)
```

Performing various visualizations given below for the data after performing aggregation (after solving Q3). 

1. Bar Plot

```python

nationality_height = df[['Nationality', 'Height(Inch)']]
average_height_by_nationality = nationality_height.groupby('Nationality')['Height(Inch)'].mean().reset_index()
average_height_by_nationality = average_height_by_nationality.sort_values(by='Height(Inch)', ascending=False)


plt.figure(figsize=(20, 10))
plt.bar(average_height_by_nationality['Nationality'], average_height_by_nationality['Height(Inch)'])
plt.xticks(rotation=90)
plt.xlabel('Nationality')
plt.ylabel('Average Height')
plt.title('Average Height by Nationality')

plt.show()
```

here we observe that player from South Sudan has the highest avg Height while that of Indonesia has the lowest avg height

```python
2. Line Plot
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['HeadingAccuracy'], label='HeadingAccuracy', linestyle='-', marker='o')
plt.plot(df['Jumping'], label='Jumping', linestyle='-', marker='o')
plt.title('HeadingAccuracy vs. Jumping Line Plot')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

4. Box Plot

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Overall', y='Value', data=df, palette='viridis')
plt.xlabel('Overall Rating')
plt.ylabel('Value')
plt.title('Box Plot of Value vs Overall Rating')
plt.xticks(rotation=90)
plt.show()
```

it reaffirms the notion that higher the overall rating higher is the value. But it is important to note that there is also a lot of variation in value, and that some players are valued more than their overall rating would suggest.

Violin PLot

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='Position', y='Overall', data=df, palette='viridis')
plt.xlabel('Position')
plt.ylabel('Overall Rating')
plt.title('Violin Plot of Overall Rating by Position')
plt.xticks(rotation=90)
plt.show()






```

the violin plot shows that there is a lot of variation in overall rating within each position. However, it also suggests that CBs and STs are generally higher-rated players than RWBs and LWBs. The density of the data is highest for the CB and ST positions, and lowest for the RWB and LWB positions. This means that there are more players with overall ratings close to the median in the CB and ST positions, and fewer players with overall ratings close to the median in the RWB and LWB positions.
